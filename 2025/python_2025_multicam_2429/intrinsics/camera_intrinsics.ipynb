```python
import json
import pandas as pd
import glob, os
import cv2
import numpy as np
```

```python
def get_files():
    json_dir = r'C:\CJH\python\FRC\vision\2025\python_2025_multicam_2429\intrinsics'
    pattern = os.path.join(json_dir, "**", "*.json")  # Recursive pattern
    json_files = glob.glob(pattern, recursive=True)
    return json_files
```

```python
def get_data(file):
    with open(file, 'r') as f:
            data = json.load(f)
            fx = data['camera_matrix'][0][0]
            cx = data['camera_matrix'][0][2]
            fy = data['camera_matrix'][1][1]
            cy = data['camera_matrix'][1][2]
            fx, fy, cx, cy
            data['fx'] = fx
            data['fy'] = fy
            data['cx'] = cx
            data['cy'] = cy
            data['d1'] = data['distortion_coefficients'][0]
            data['d2'] = data['distortion_coefficients'][1]
            data['d3'] = data['distortion_coefficients'][2]
            data['d4'] = data['distortion_coefficients'][3]
            data['d5'] = data['distortion_coefficients'][4]
            data['x_res'] = data['img_size'][0]
            data['y_res'] = data['img_size'][1]
    return data
```

```python
json_files = get_files()
data = [ get_data(f) for f in json_files]
df = pd.DataFrame(data)
df;
```

```python
def get_average_intrinsics(cam, scale=1.0):
    gb = df.groupby("camera").mean(numeric_only=True)
    gb= gb.reset_index()
    gb.columns
    df_temp = gb[gb['camera'] == cam]
    #cam = 'HD Pro Webcam C920 (046d:08e5)'
    fx = df_temp['fx'].iloc[0].item() * scale
    fy = df_temp['fy'].iloc[0].item() * scale
    cx = df_temp['cx'].iloc[0].item() * scale
    cy = df_temp['cy'].iloc[0].item() * scale
    camera_matrix = np.array([
        [fx, 0, cx],  # fx,  0, cx
        [0, fy, cy],  #  0, fy, cy
        [0, 0, 1]            #  0,  0, 1
        ], dtype=np.float32)
    distortions= np.array([df_temp['d1'].iloc[0].item(),
                  df_temp['d2'].iloc[0].item(), df_temp['d3'].iloc[0].item(), 
                  df_temp['d4'].iloc[0].item(), df_temp['d5'].iloc[0].item(),],
                  dtype=np.float32)
    dstring = f"{{'fx':{fx:.2f}, 'fy':{fy:.2f}, 'cx':{cx:.2f}, 'cy':{cy:.2f} }}"
    dmesg = f""
    return dstring, [fx, fy, cx, cy], camera_matrix, distortions
```

```python
def convert_intrinsics(x_res,y_res, cam_json, use_max=True):
    full_x, full_y = cam_json['x_res'], cam_json['y_res']
    fx_full, fy_full = cam_json['fx'], cam_json['fy']
    cx_full, cy_full = cam_json['cx'], cam_json['cy']
    scale_x = x_res / full_x
    scale_y = y_res / full_y

    if use_max:
        scale= max(scale_x, scale_y)
        fx = scale * fx_full
        fy = scale * fy_full
    else:
        fx = scale_x * fx_full
        fy = scale_y * fy_full
    cx = scale_x * cx_full
    cy = scale_y * cy_full
    
    return {'fx':fx, 'fy':fy, 'cx':cx, 'cy':cy}, 
```
